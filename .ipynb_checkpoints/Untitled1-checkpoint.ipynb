```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
label = ["设备数据","周波数据","谐波数据"]
data1 = [] #设备数据
for i in range(11):
    data1.append(pd.read_excel("A题部分数据\\附件1\\YD%d.xlsx"%(i+1),sheet_name = label[0]))
data2 = [] #周波数据
for i in range(11):
    temp = pd.read_excel("A题部分数据\\附件1\\YD%d.xlsx"%(i+1),sheet_name = label[1])
    del temp["time"]
    data2.append(temp)
data3 = [] #谐波数据
for i in range(11):
    temp = pd.read_excel("A题部分数据\\附件1\\YD%d.xlsx"%(i+1),sheet_name = label[2])
    del temp["time"]
    data3.append(temp)

```

```python
plt.figure(figsize = (12,8))

plt.plot(data1[9]["PFC"])
#plt.plot(data[5]["PFC"])

#plt.set_title('IC',fontsize=12,color='r')  
plt.title("QC\PC",fontsize=17)
#plt.legend([u"白炽灯",u"节能灯"], loc = 0, ncol = 2,fontsize=17)  
plt.show()
```

```python
plt.figure(figsize = (20,16))
k = 0
for i in data2:
    temp = i.iloc[:,1]
    if k+1 != 10:
        k +=1
        continue
    else:
        pass
    #plt.subplot(3,4,k+1)
    plt.plot(temp)
    plt.scatter(range(len(temp)),temp,c = 'r',marker = 'o')  

    k +=1
    plt.title("YD%d"%k,fontsize=17)  
plt.show()
    
```
